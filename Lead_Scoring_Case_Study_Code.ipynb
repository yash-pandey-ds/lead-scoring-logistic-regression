# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=21)
```

```python
X.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
features = ['TotalVisits','Total Time Spent on Website','Page Views Per Visit']
```

```python
X_train[features] = scaler.fit_transform(X_train[features])
X_test[features] = scaler.transform(X_test[features])
```

# Modeling

```python
import statsmodels.api as sm
```

```python
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

logm = sm.GLM(y_train,X_train_sm,family=sm.families.Binomial()).fit()
logm.summary()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve,auc, precision_recall_curve
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_test_prob = model.predict_proba(X_test)[:,1]
y_test_prob
```

setting threshold 

```python
threshold = 0.4
y_pred_test = (y_test_prob >= threshold).astype(int)
y_pred_test
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)
```

```python
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], 'k--', color='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
print(f'The Accuracy of the model is: {accuracy_score(y_test,y_pred_test)*100}%')
```

```python
threshold_list = [0.30, 0.40, 0.50, 0.60]
for i in threshold_list:
    y_test_prob = model.predict_proba(X_test)[:,1]
    y_train_prob = model.predict_proba(X_train)[:,1]
    
    y_pred_test = (y_test_prob >= i).astype(int)
    y_pred_train = (y_train_prob >= i).astype(int)
    
    print(f'For Threshold: {i} :')
    print()
    print(f'Train Accuracy: {round(accuracy_score(y_train, y_pred_train)*100,2)}')
    print(f'Train Precision: {round(precision_score(y_train, y_pred_train)*100,2)}')
    print(f'Train Recall: {round(recall_score(y_train, y_pred_train)*100,2)}')
    print()
    print(f'Test Accuracy: {round(accuracy_score(y_test, y_pred_test)*100,2)}')
    print(f'Test Precision: {round(precision_score(y_test, y_pred_test)*100,2)}')
    print(f'Test Recall: {round(recall_score(y_test, y_pred_test)*100,2)}')
    print('-')
    print('-----------------------------------------')
    print('-')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
confusion_matrix = confusion_matrix(y_train, y_pred_train)
print(confusion_matrix)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['Not Converted', 'Converted'])
```

```python
cm_display.plot()
plt.show()
```

```python
TP = confusion_matrix[1,1] #True Positives
TN = confusion_matrix[0,0] #True Negatives
FP = confusion_matrix[0,1] #False Positives
FN = confusion_matrix[1,0] #False Negatives
```

```python
sensitivity = round(TP/float(TP + FN),2)
print(sensitivity)
```

```python
specificity = round(TN/float(TN + FP),2)
print(specificity)
```
